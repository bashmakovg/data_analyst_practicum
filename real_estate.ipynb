Здесь я решил сделать сводную таблицу, в которой мы увидим медианную долю площади кухни и жилой зоны от общей площади квартиры. В дальнейшем при заполнении пропущенных значений можно пользоваться значениями этой сводной таблицы

```python
pivot_area = data.pivot_table(index='rooms', values=['total_area', 'living_area', 'kitchen_area'], aggfunc='median' ).reset_index()
pivot_area['dolya_kuhni'] = pivot_area['kitchen_area']/pivot_area['total_area']
pivot_area['dolya_zhiloy'] = pivot_area['living_area']/pivot_area['total_area']
pivot_area
```

```python
# здесь я сделал срез, чтобы посмотреть в каких квартирах у нас пропущены значения площади кухни
data.query('kitchen_area.isna()')['rooms'].value_counts()
```

А далее можно сделать цикл, который пробегается по значениям квартир от 1 до 9, и если в этой строке отсутствует значение площади кухни, то оно будет заполнено медианным значением от общей площади квартиры

```python
for i in range(1, 10):
    data.loc[(data['rooms']==i)&(data['kitchen_area'].isna()), 'kitchen_area']=(data['total_area']*pivot_area['dolya_kuhni'][i])
#здесь мы заменяем пропущенные значения площади кухни в квартирах студиях
```

```python
# выполняю проверку на пропущенные значения в этой колонке
data['kitchen_area'].isna().sum()
```

```python
for k in range(0, 8):
    data.loc[(data['rooms']==k)&(data['living_area'].isna()), 'living_area']=(data['total_area']*pivot_area['dolya_zhiloy'][k])
#здесь мы заменяем пропущенные значения площади кухни в квартирах студиях
```

```python
# 8-ми комнатные квартиры выпали из этого списка, поэтому для квартир с 9 комнатами выполним отдельное действие
data.loc[(data['rooms']==9)&(data['living_area'].isna()), 'living_area']=(data['total_area']*pivot_area['dolya_zhiloy'][9])
```

### Добавляем категорию этажности квартиры

Мы должны создать столбец, в который исходя из значения этажа будет попадать значение - первый, последний или другой.
мы можем это реализовать с помощью функции, которая принимает на вход весь датасет, а потом сравнивает значение в колонке, и если оно равно 1, то это первый этаж, если равно количеству этажей в доме, то последний. Если не удовлетворяет первым двум условиям то значение попадает в категорию 'другой'.

```python
#создаем функцию, принимающую весь датафрейм
def floor_categ(data):
    if data['floor']==1:
        return 'первый'
    if data['floor']==data['floors_total']:
        return 'последний'
    return 'другой'
```

```python
# применяем нашу функцию
data['floor_category'] = data.apply(floor_categ, axis=1)
#axis=1 означает то, что перебираем не строки а столбцы
```

```python
data.loc[:, 'floor':'floor_category'].head()
```

- Такому рынку тоже присуща сезонность, в этом мы убедились опираясь на полученные графики, которые показывают, что в летний период (июнь) квартиру можно купить дешевле почти на 14% по сравнению с ее пиковой стоимостью весной, либо дождаться предновогодних акций. 
- Кризис 2014 года хорошо ударил по цене, но уже в 2017 году после небольшой коррекции мы наблюдаем постоянный рост недвижимости в цене.

В общем повезло тем, кто в 2016 году успел взять себе 3х комнатную квартиру около центра Санкт-Петербурга с хорошими высокими потолками. Сейчас он либо наслаждается жизнью в ней, либо выгодно продал и вложил в свое дело.
